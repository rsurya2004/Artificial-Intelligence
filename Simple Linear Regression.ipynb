# Simple Linear Regression

# Importing The Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing The Dataset 

```python
dataset = pd.read_csv('C:\\Users\\YOGMAYA\\Desktop\\Datasset\\Salary_Data.csv')
x = dataset.iloc[ : , :-1].values # .values = array
y = dataset.iloc[ : , -1].values
```

# Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split #Funtion
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state = 0) #1row at a time = random_state
```

# Training the Simple Linear Regression Model on the Training Data

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train , y_train)
```

# Predicting the Test set Result

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
print(r2_score (y_test , y_pred))
```

# Visualising the Training set Results

```python
plt.scatter(x_train , y_train , color ='red')
plt.plot( x_train , reg.predict(x_train) , color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()
```

# Visualising the Test set Results

```python
plt.scatter(x_test , y_test , color = 'blue' )
plt.plot(x_train , reg.predict(x_train) , color = 'red')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
