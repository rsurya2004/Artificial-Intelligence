# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# Importing the Dataset

```python
df = pd.read_csv('C:\\Users\\YOGMAYA\\Desktop\\Datasset\\Social_Network_Ads.csv')
x = df.iloc[ : , :-1].values
y = df.iloc[ : , -1].values
df
```

# Slitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split( x , y , test_size = 0.2 , random_state = 0 )
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

# Training KNN model on the Training Set

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier( n_neighbors = 5 , metric = 'minkowski')  # minkowski distance
classifier.fit(x_train , y_train)
```

# Predicting the Test set results

```python
y_pred = classifier.predict(x_test)
print( np.concatenate((y_pred.reshape(len(y_pred),1) , y_test.reshape(len(y_test),1))  ,1))
```

# Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test ,y_pred)
```
