# 18. Find min of Total Deposits and max of Estimated Loss based on Charter Type and Fail Year where Fail_Year> 2010 and Institution Name is not First State Bank .Depict the result on a stacked bar chart(unstacking is not required)

```python
where18 = df[(df['Fail_Year'] > 2010) & (df['Institution Name'] != 'First State Bank')]
where18
```

```python
q18_1 = where18.groupby(['Fail_Year' , 'Charter Type'])['Total Deposits'].min()
q18_2 = where18.groupby(['Fail_Year' , 'Charter Type'])['Estimated Loss (2015)'].max()
q18 = pd.concat([q18_1,q18_2] , axis = 1)
q18
```

```python
q18 = q18.reset_index()  # Multi Index ---> Columns
q18
```

```python
# Combine the integer and string columns into a single categorical column
q18['CombinedCategory18'] = q18['Fail_Year'].astype(str) + ' - ' + q18['Charter Type']
q18
```

```python
#Stacked Bar Chart
plt.figure(figsize=(25, 25))
plt.grid()
plt.bar(q18['CombinedCategory18'], q18['Total Deposits'], color='red' )
plt.bar(q18['CombinedCategory18'], q18['Estimated Loss (2015)'], color='blue' , bottom = q18['Total Deposits'] )
plt.title('Min of Total Deposits and Max of Estimated Loss based on Charter Type and Fail Year')
plt.xlabel('Charter Type and Fail Year')
plt.ylabel('Min of Total Deposits and Max of Estimated Loss')
plt.show()
```
