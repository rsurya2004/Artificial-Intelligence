# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
df
```

# BarGraph 

```python
mt = df['Market'].value_counts()
mt
```

```python
plt.figure(figsize=(10,10))
plt.bar(mt.index, mt.values, color='pink')
plt.grid()
plt.xlabel('Market')
plt.ylabel('Counts')
plt.title('Market Types vs Its Counts')
plt.show()
```

# Line Graph

```python
plt.figure(figsize=(10,15))
plt.plot(mt.index , mt.values, color='blue' , marker='+')
plt.grid()
plt.xlabel('Market')
plt.ylabel('Counts')
plt.title('Market Types vs Its Counts')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.pie(mt, labels = mt.index, autopct= '%.2f%%')
plt.legend(loc=1)
plt.title('Market and its counts')
plt.show()
```

# Scatter Chart

```python
df.groupby('Market')['Profit'].sum()
```

```python
p = df[df['Product'] == 'Green Tea']
plt.figure(figsize=(15,15))
plt.scatter(p['Sales'], p['Profit'], color='red', marker='+')
plt.grid()
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# Histogram

```python
te = df['Total Expenses']
plt.figure(figsize=(15,15))
plt.hist(te, color='green')
plt.grid()
plt.show()
```
